## ***2.4 Data Cleaning***

To start our data cleaning process, we begin by checking for duplicate entries in the dataset, since they are redundant and do not contribute to our analysis. We will identify and remove one of the duplicate records to ensure data integrity.

```python
# We find and display the duplicate rows in the data

CDC_data[CDC_data.duplicated(keep = False)].sort_values(["Weight_(kg)", "General_Health"])
```

```python
# We calculate and print the number of duplicate rows in the data

duplicates_number = CDC_data.duplicated().sum()

print("Number of duplicates:", duplicates_number)
```

```python
# We remove the duplicate rows in the data

CDC_data.drop_duplicates(inplace = True)
CDC_data
```

While conducting our data analysis, we observed that the "Diabetes" column contains four unique values. To simplify our analysis and make it more straightforward, we will simplify these values by categorizing them as 'No' and 'Yes'.

```python
# Transforming the "Diabetes" column

CDC_data["Diabetes"] = CDC_data["Diabetes"].str[:3].str.replace(',', '')
for i in categorical_columns:
  print("{0:15s} shape={1}".format(i, CDC_data[i].unique()))
```

With all columns now containing only 'yes' and 'no' values, we can perform binary encoding, assigning 'yes' as 1 and 'no' as 0. This binary encoding simplifies our data and facilitates easier plotting and analysis.

```python
# We identify the binary columns in the data, excluding 'Diabetes'

binary_columns = CDC_data.columns[CDC_data.isin(['Yes', 'No']).any()].tolist()
binary_columns
```

```python
# We binary encode the binary columns of our data

for i in binary_columns:
  CDC_data[i] = CDC_data[i].map(dict(Yes=1, No=0))
CDC_data.head()
```

For the column "General_Health", we can employ ordinal encoding, which involves assigning ordinal values based on a scale. In this case, we can use the following scale: {'Poor': 0, 'Fair': 1, 'Good': 2, 'Very Good': 3, 'Excellent': 4}. This encoding allows us to represent the varying levels of general health numerically.

```python
# We perform ordinal encoding for the "General_Health" column, mapping categorical values to numerical values

CDC_data["General_Health"] = CDC_data["General_Health"].map(dict({'Poor': 0, 'Fair': 1, 'Good': 2, 'Very Good': 3, 'Excellent': 4}))
CDC_data.head()
```

To encode the "Age_Category" column, we can apply ordinal encoding. In this instance, we can utilize the following scale: {"18-24": 0, "25-29": 1, "30-34": 2}. This encoding enables us to numerically represent the different age groups. It is particularly pertinent because age has a progressively increasing impact on the risk of cardiovascular diseases (CVDs). Therefore, opting for ordinal encoding is a meaningful choice over one-hot encoding in this context.

```python
CDC_data["Age_Category"] = CDC_data["Age_Category"].map(dict({"18-24": 0, "25-29": 1, "30-34": 2}))
```

```python
CDC_data["Age_Category"].value_counts()
```

When looking at the data, we've observed an unusually high BMI value of 96.52, which appears to be an outlier. To verify the consistency of this column, we can cross-reference it with the BMI calculation formula. The BMI formula is typically calculated as follows:

$BMI = \frac{{\text{weight in kilograms}}}{{(\text{height in meters})^2}}$

By using this formula and comparing it to the values in the column, we can ensure data consistency and identify any potential errors.

```python
# We calculate "Predicted_bmi" from our data based on weight and height

CDC_data["Predicted_bmi"] = CDC_data["Weight_(kg)"] /((CDC_data["Height_(cm)"]/100)**2)
CDC_data.describe()
```

```python
# Check the outlier data of BMI (BMI=96.52)
CDC_data.loc[CDC_data["BMI"] == 96.520000]
```

```python
# We locate the rows where the difference between "BMI" and "Predicted_bmi" is greater than 0.5

CDC_data.loc[(CDC_data["BMI"] - CDC_data["Predicted_bmi"]) > 0.5]
```

We've observed that only one entry deviates from our expected value by slightly more than 0.5, but it's not significantly different. Therefore, we'll retain this entry and won't drop any values from the dataset.

We will try to determine an individual's BMI category (underweight, normal, overweight, or obese) by leveraging the BMI column. This decision is supported by the fact that the dataset's calculated values closely align with those derived from the BMI formula. Subsequently, we will exclude the height, weight, and BMI columns, as they are deemed redundant for our research objectives. \\
We will be using the following ranges obtained form the CDC: \\
If your BMI is less than 18.5, it falls within the underweight range. \\
If your BMI is 18.5 to 24.9, it falls within the Healthy Weight range. \\
If your BMI is 25.0 to 29.9, it falls within the overweight range. \\
If your BMI is 30.0 or higher, it falls within the obese range. \\


```python
percentage_underweight = (((CDC_data["BMI"] > -float("inf")) & (CDC_data["BMI"] <= 18.50)).sum() / len(CDC_data)) * 100
percentage_normal = (((CDC_data["BMI"] > 18.50) & (CDC_data["BMI"] <= 25)).sum() / len(CDC_data)) * 100
percentage_overweight = (((CDC_data["BMI"] > 25) & (CDC_data["BMI"] <= 29.99)).sum() / len(CDC_data)) * 100
percentage_obese = (((CDC_data["BMI"] > 29.99) & (CDC_data["BMI"] <= float("inf"))).sum() / len(CDC_data)) * 100
```

```python
CDC_data["BMI_Category"] = pd.cut(
    CDC_data["BMI"],
    bins=[-float("inf"), 18.5, 25, 29.99, float("inf")],
    labels=["underweight", "normal", "overweight", "obese"]
)
```

```python
round(CDC_data["BMI_Category"].value_counts(normalize=True)*100, 2)
```

```python
print("Normal:", round(percentage_normal, 2))
print("Overweight:", round(percentage_overweight, 2))
print("Obese:", round(percentage_obese, 2))
print("Underweight:", round(percentage_underweight, 2))
```

##***3.1 Does having depression affect the probability of getting heart disease if one has arthritis or diabetes already?***


From the given data, we would like to find out whether having depression increases one's probability of getting heart disease given the chance that the person is diagnosed with arthritis or diabetes.

To answer this question, we will be using the following variables: Arthritis, Depression, Diabetes, and Sex. The specific reasons for selecting these variables are as follows: both arthritis and diabetes increase one's probability of getting heart disease, as arthritis implies inflammation within the human body, weakening blood vessels; diabetes indicates high blood sugar within the blood vessels, also weakening them. Additionally, we considered that having arthritis or diabetes can affect one's mindset and therefore increase the chance of having depression. We also chose to plot for sex since we found significant differences in probability between the two genders. Lastly, since we were curious about whether depression weakens one's body, making it more vulnerable to heart disease, we will plot it on our x-axis.

```python
sns.catplot(kind = "bar", x = "Depression", y = "Heart_Disease", data = CDC_data, hue = "Sex", col = "Arthritis", row = "Diabetes", palette = "pastel");
plt.suptitle("Probability of having a Heart Disease by Depression, Diabetes, Sex and Arthritis", y=1.05)
plt.show();
```

From the graph above, we can tell that there's no significant difference between having depression or not when the person has diabetes only or doesn't have any. On the other hand, the probability slightly increases if one has arthritis and depression at the same time. However, the bottom-right plot shows us an interesting fact that both the probability increases despite having depression or not, indicating that diabetes and arthritis do affect the likelihood of getting heart disease. What's even more fascinating is that people tend to be two times more likely to have heart disease if they have depression compared to those who don't.

Thus, our conclusion is that depression does influence the probability of having heart disease, though it's not obvious when one doesn't have other diseases but is especially significant when you have both arthritis and diabetes.

##***3.2 Does exercise influence the probability of getting heart disease in individuals who already have conditions like arthritis or other cancers?***



In the first question, we observed that depression, diabetes, and arthritis indeed have an impact on the probability of having heart disease. Consequently, we aim to conduct further research on how to reduce the risk of heart disease in individuals who may be dealing with less favorable health conditions(such as having cancer or arthritis).

To explore this question, we will use the following variables: Other_cancer, Arthritis, Exercise, and Sex. Upon calculating a heatmap of the correlation matrix for the numeric columns of the dataset, we have identified that the variables "Other_cancer" and "Arthritis" exhibit relatively higher correlation values with heart disease, namely 0.06 and 0.07, respectively. Therefore, we have chosen these two representative disease variables, "Other_cancer" and "Arthritis," for further analysis.

We want to know if engaging in doing exercise can lead to improved cardiovascular health(CVH), even when you have other medical conditions(like other cancer and arthritis). Does exercise reduce the risk of heart disease? The chart below illustrates our findings.

```python
sns.catplot(kind = "bar", x = "Exercise", y = "Heart_Disease", data = CDC_data, hue = "Sex", col = "Arthritis", row = "Other_Cancer", palette = "pastel");
plt.suptitle("Probability of having a Heart Disease by Exercise, Other Cancer, Sex and Arthritis", y=1.05)
plt.show();
```

From the chart above, we can see that when a person does not have other cancer or arthritis, there is no significant difference in having the risk of heart disease, whether they engage in exercise or not. Moreover, when an individual has either other cancer or arthritis as their single disease condition, their risk of heart disease is slightly higher compared to those without other cancer and arthritis. Exercise does not have a significant impact on the probability of developing heart disease in this scenario; it only provides a modest reduction in the risk of having heart disease.

However, a noteworthy discovery emerges when individuals have both cancer and arthritis(the subplot in the lower right corner). In such cases, exercise plays a vital role in significantly lowering the risk of heart disease, particularly among males, where it more than halves the probability.

In conclusion, exercise appears to be particularly beneficial for individuals dealing with multiple health conditions, effectively reducing the risk of heart disease, especially in cases of concurrent cancer and arthritis for males.

***Conclusion:***

# 4. Modeling

## Logistic Regression Normalized

We begin the modelling process by applying the simplest model in our toolset, which happens to be logistic regression. However, before we can effectively apply this model, we must guarantee that all of our data is in numerical format. As a result, we will conduct one-hot encoding on the remaining categorical columns that we previously kept for visualisation. We will also standardise the float columns in order to bring them to a consistent scale and better understand their impact on the modelling process. To compensate for the uneven distribution of the target variable, we shall set the class weights of the model to "balanced".

```python
# We import necessary libraries for data modeling

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
import xgboost as xgb
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.metrics import roc_curve, auc, roc_auc_score
from sklearn.ensemble import VotingClassifier
```

In the one-hot encoding step, we will exclude the "normal" and "never checkup" columns to maintain them as the reference baseline for our model.

```python
# We one hot encode the columns with categorical data

CDC_data = pd.get_dummies(CDC_data, columns=["Sex"], drop_first=True)
CDC_data = pd.get_dummies(CDC_data, columns=["Checkup"])
CDC_data = pd.get_dummies(CDC_data, columns=['BMI_Category'], prefix='BMI')
CDC_data.drop(["BMI_normal", "Checkup_Never"], axis=1, inplace=True)
```

We will split the dataset for modeling. We create the variable X by excluding the "Heart_Disease" column, which contains our independent variables, and we store the target variable, "Heart_Disease," in the variable y. This allows us to set up our data for a supervised learning task where we'll use the features in X to predict the target variable in y.

```python
# We create the independent variables (X) and dependent variable (y) for data modleing

X = CDC_data.drop(["Heart_Disease"], axis=1)
y = CDC_data["Heart_Disease"]
```

We're adjusting specific columns in our dataset X to have a consistent scale for machine learning. These columns, which include "Alcohol_Consumption," "Fruit_Consumption," "Green_Vegetables_Consumption," and "FriedPotato_Consumption," have their values modified to have a mean of 0 and a standard deviation of 1. Because of this standardisation, they are more compatible with various machine learning techniques that are sensitive to feature scaling. This adjusted dataset is then saved as X_norm, which is a copy of the original X with the selected columns standardised for modelling.

```python
# We use Standardscaler to normalize data for certain columns

scaler = StandardScaler()
column_to_norm = ["Alcohol_Consumption", "Fruit_Consumption", "Green_Vegetables_Consumption", "FriedPotato_Consumption"]
X[column_to_norm] = scaler.fit_transform(X[column_to_norm])
X_norm = X.copy()
X_norm
```

We use the train_test_split function from scikit-learn to split the data into training data and testing data.

```python
# We split our input data into 60% training data and 40% testing data

X_train, X_test, y_train, y_test = train_test_split(X_norm, y, test_size=0.4, random_state=327, stratify = y)
```

We configure the logistic regression model to automatically adjust class weights, specifically using the "balanced" option, which helps address imbalances in the target variable distribution. Subsequently, we fit the model to the training data with the fit method, taking X_train as the input features and y_train as the target variable. This fitting process enables the model to acquire knowledge from the training dataset, which it will subsequently utilize to make predictions on our testing data.

```python
model = LogisticRegression(class_weight= "balanced")
model.fit(X_train, y_train)
```

We create binary classification predictions using the probabilities generated by a logistic regression model with a threshold of 0.7 to strike a balance between precision and recall in our classification task. A threshold of 0.7 is chosen to make our predictions more conservative, ensuring that an instance is classified as positive only when the model is relatively confident.

```python
y_pred = (model.predict_proba(X_test)[:,1] >= 0.7).astype(bool)
```

We calculate accuracy to measure the proportion of correct predictions, generate a confusion matrix for a detailed breakdown of prediction outcomes, and create a classification report with metrics like precision, recall, F1-score, and support for each class.

```python
# We print the accuracy score, confusion matrix and classification report to view the performance

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print("Accuracy of the model:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", class_report)
```

We visualize the feature importances of a logistic regression model by calculating their coefficient from the model.

```python
feature_importances = model.coef_[0]
abs_feature_importances = np.abs(feature_importances)
sorted_feature_indices = np.argsort(abs_feature_importances)[::-1]
all_feature_names = X_train.columns[sorted_feature_indices]

plt.figure(figsize=(10, 8))
plt.barh(all_feature_names, feature_importances[sorted_feature_indices], color = "green")
plt.xlabel("Coefficient Value")
plt.title("All Features in Logistic Regression Model")
plt.gca().invert_yaxis()
plt.show()
```

##Voting Classifier

Given our extensive range of constructed models, we can use ensemble modeling with all models except K-Nearest Neighbors. We want to utlize the collective predictive power of these diverse models to maximize the variance captured by our final model.

```python
# We specify the five models that are using in the voting classifier

clf1 = LogisticRegression(class_weight= "balanced")
clf2 = DecisionTreeClassifier(max_depth=3, class_weight="balanced", random_state=327)
clf3 = RandomForestClassifier(n_estimators=1000, class_weight="balanced", max_depth=3, random_state=327)
clf4 = xgb.XGBClassifier(objective='binary:logistic', scale_pos_weight=100,  max_depth=3, learning_rate=0.1, n_estimators=1000, subsample=0.8, colsample_bytree=0.8, random_state=327)
clf5 = GaussianNB()

# We implement the voting classifier method that takes the five results from our models and output their average as the final probability

voting_clf = VotingClassifier(estimators=[('lr', clf1), ('dt', clf2), ('rf', clf3), ("xg", clf4), ("gnb", clf5)], voting='soft')
voting_clf.fit(X_train, y_train)
```

```python
y_pred = (voting_clf.predict_proba(X_test)[:,1] >= 0.7).astype(bool)
```

```python
# We print the accuracy score, confusion matrix and classification report to view the performance

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print("Accuracy of the model:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", class_report)
```

We can visualize the confusion matrix by rendering it as a heatmap, which allows us to gain a more comprehensive understanding of the distribution of predictions.

```python
#We create and display a heatmap of the confusion matrix

sns.heatmap(conf_matrix, cmap="YlGnBu", annot=True, fmt='d', annot_kws={"size": 12})
plt.xlabel("Predicted Labels", fontsize=12)
plt.ylabel("True Labels", fontsize=12)
plt.title("Confusion Matrix", fontsize=16)
plt.show()
```

In order to assess our model's performance, we can create a Receiver Operating Characteristic curve. This visual representation allows us to evaluate the trade-off between false positive rate (FPR) and true positive rate (TPR) across different classification thresholds.

```python
# We compute ROC curve, AUC and plot it

y_probs = voting_clf.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_probs)
roc_auc = roc_auc_score(y_test, y_probs)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (AUC = {:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], color='black', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```

We calculate the Youden's Index, which is a useful metric for determining the optimal threshold in binary classification.

```python
# We calculate the Youden Index and find the best threshold

youden_index = tpr - fpr
best_threshold = thresholds[np.argmax(youden_index)]
best_threshold
```
